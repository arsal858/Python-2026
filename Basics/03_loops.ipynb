**What are Loops**  
Loops are the powerful way of repeating a piece of code again and again.

```python
i=3
while i!=0:
    print('hy')
    i=i-1
```

```python
i=0
while i<=3:
    print('buy')
    i+=1
```

```python
# for loop
for i in [1,2,3]:
    print('Meou')
```

```python
for i in range(10):
    print('Arslan')
```

```python
n=int(input('What is n: '))
if n<0:
    n=int(input('What is n:'))
else:
    print('Your number is Positive')
```

```python
while True:
    n=int(input('Please enter n: '))
    if n<0:
        continue
    else: 
        break
for _ in range(n):
    print('Arslan')
```
