```python
# 2d array
import numpy as np
arry=np.array([[1,3,4],[5,6,8]])
print(arry)
print(np.ndim(arry))
print(type(arry))
```

```python
# 3d array
array_3d=np.array([[[1,3,5],[7,7,8],[1,1,1]]])
print(array_3d)
print(np.ndim(array_3d))
print(type(array_3d))
```

```python
# multidiemensional array
array_mul_d=np.array([[[[1,4,5],[1,7,9],[8,9,10],[1,1,5]]]])
print(array_mul_d)
print(np.ndim(array_mul_d))
print(type(array_mul_d))
```

```python
# dynamic data types of array
array=np.array([[[1,3,4],[1,3,4]]],dtype=float)
array_1=np.array([[[1,0,-1],[1,0,-8]]],dtype=bool)
array_2=np.array([[1,9,10],[1,4,6]],dtype=complex)
print(array_1)
print(array)
print(array_2)
print(np.ndim(array_2))
```

```python
# array range function
array=np.arange(1,51,2)
print(array)
# reshape
array_1=np.arange(1,41).reshape(8,5)
print(array_1)
print(np.ndim(array_1))
```

```python
# random
np.random.random((1,10))
```

```python
# linspace
np.linspace(1,5)
```

```python
# identity
np.identity(3)
```

Array Attributes

```python
a=np.arange(1,15)
b=np.arange(12,dtype=float).reshape(3,4)
c=np.arange(8).reshape(2,2,2)
print(a)
print(b)
print(c)
```

```python
# scaller operations
a=np.arange(1,13).reshape(4,3)
print(a)
print('='*15)
print(a*3)
```

```python
a=np.array([[1,5,7],[5,8,9]])
b=np.array([[4,8,16],[12,16,18]])

print(a)
print(b)
print(np.ndim(a))
print('='*20)

dot=np.dot(a,b)
print(dot)
```
