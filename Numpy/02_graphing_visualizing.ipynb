```python
# linear function
import numpy as np 
import matplotlib.pyplot as plt
x=np.linspace(-10,10)
y=x
plt.plot(x,y,'k')
plt.show()
```

```python
# s
x=np.linspace(-5,5)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
# sin
x=np.linspace(-15,15)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# sifmoid
x=np.linspace(-9,9,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
plt.show()
```

```python
# log function
x=np.linspace(-10,10,100)
y=np.log(x)
plt.plot(x,y)
plt.show()
```
